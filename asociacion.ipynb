```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('orders.csv', sep=";")
print(df)
```

```python
data = list(df["product_name"].apply(lambda x:x.split(';')))
```

```python
df= df.groupby('product_name').filter(lambda x : len(x)>400)
```

```python
df_encoded= pd.get_dummies(df['product_name'])
df_grouped = df_encoded.groupby(df['order_id']).sum()
df_grouped.to_csv('datos2.csv', sep=";")
df_grouped
```

```python
df_grouped.sum().sort_values(ascending=False).to_csv('soporteOrder.csv', sep=";")
```

```python
rules=apriori(df_grouped, min_support=0.01, use_colnames=True)
rules['length'] = rules['itemsets'].apply(lambda x: len(x))
sorted_rules = rules.sort_values('support', ascending = False)
sorted_rules
```

```python
# printing the frequntly items 
rules[ (rules['length'] == 2) &
                   (rules['support'] >= 0.01)]
sorted_rules = rules.sort_values('support', ascending=False)
sorted_rules
```

```python
rules[ (rules['length'] == 3)]
sorted_rules1= rules.sort_values('support', ascending=False)
sorted_rules1
```

```python
#tabla de reglas de asociacion
reglasAsociacion = association_rules(rules, metric="confidence", min_threshold=0.3)
reglasAsociacion["antecedents_length"] = reglasAsociacion["antecedents"].apply(lambda x: len(x))
reglasAsociacion["consequents_length"] = reglasAsociacion["consequents"].apply(lambda x: len(x))
reglasAsociacion=reglasAsociacion.sort_values("confidence",ascending=False)
reglasAsociacion.to_csv('reglasAsociacionSoporte.csv', sep=";")
reglasAsociacion
```

```python
import pandas as pd
import arff
```

```python
df = pd.read_csv('Data.csv')
df
```

```python
#Change the 0 and 1 to False and True
df = df.replace({0: "0", 1: "1"})
df
```

```python
# df to arff
data = {
    'description': u'',
    'relation': 'relation',
    'attributes': [(col, df[col].unique().tolist()) for col in df.columns],
    'data': df.values
}

# Write
with open('yourfile.arff', 'w') as f:
    arff.dump(data, f)
```

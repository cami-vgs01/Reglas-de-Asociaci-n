```python
import pandas as pd
import arff
```

```python
df = pd.read_csv('datos2.csv', sep=';', dtype='str')
df
```

```python
# df to arff
data = {
    'description': u'',
    'relation': 'relation',
    'attributes': [(col, df[col].unique().tolist()) for col in df.columns],
    'data': df.values
}

# Write
with open('yourfile.arff', 'w') as f:
    arff.dump(data, f)
```
